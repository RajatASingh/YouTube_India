<h2 style = "font-family:serif;"> 1. Demand of category type in India based on views:</h2>



```python
# Checking outliers In  views column group by category type:


fig = px.box(youtube, y="category_type",
             x="views" , 
             height= 1000 ,
             width= 1100,
             color= "category_type" ,
             notched=True,
            title= "figure 1.1:  Outliers View")
fig.show()
```

* We can see that there are outliers in most of the categories. These outliers can manipulate our results and can have a huge impact.

```python
#  function to remove outliers:

def is_outlier(x):
    
    q1 = np.percentile(x , 25 ,interpolation= 'midpoint')
    q3 = np.percentile(x , 75 ,interpolation= 'midpoint')
    
    iqr = q3 - q1
    
    upper = q3 + 1.5* iqr
    lower = q1 - 1.5 * iqr
    return ~x.between(lower, upper)
    
```

```python
# Storing data without outliers in a variable:

outliers = youtube[~youtube.groupby('category_type')['views'].apply(is_outlier)]
```

```python
# Graph after removing all the outliers with respect to category:
 
fig = px.box(outliers, y="category_type",
             x="views" , 
             height= 1000 ,
             width= 1100, 
             color= "category_type" ,
             notched=True,
            title= "figure 1.2: Without Outliers")
fig.show()
```

* So we have stored the views without outliers and other columns in the outliers variable. This will help us to get a somewhere accurate answer to our question. <br>
* By looking at the graph we can say that the difference between graph 1.1 and graph 1.2 is quite visible. <br>
* If we particularly talk about a single category Entertainment we can see that in figure 1.1 the top outliers are around 125 million and after removing the outliers and plotting graph 1.2  we can see the highest values are 1.4 million. Now it is easy to understand how outliers can have manipulated our result

```python
# Making a data frame which will have:
#category_type, sum of views as "views",
#video publish count, 
#sum of likes as "likes", 
#some of dislikes as "dislikes"
# per video view ratio
# per video like ratio

cat_sum_views = outliers.groupby("category_type" ).agg({"views" : "sum" , 
                                       "publish_date" : "count",
                                       "likes" : "sum", 
                                       "dislikes" : "sum"}).reset_index().rename(columns = {"publish_date" : "video_publish"})
```


<h2 style = "font-family:serif;"> 9. Find The Like / Dislike ratio of YouTube videos.</h2>



```python
fiveChannelsAnalysis = youtube.groupby("channel_title").agg({"publish_date" : "count",
                                      "views" : "sum" ,
                                      "likes" : "sum" , 
                                      "dislikes" : "sum" ,
                                      "comment_count" : "sum" }
                                    ).reset_index().rename(columns = {"publish_date" : "vid_count"})
                            
```

```python
fiveChannelsAnalysis["likes_ratio"] = round((fiveChannelsAnalysis["likes"] / fiveChannelsAnalysis["views"] ) * 100,2)

fiveChannelsAnalysis["dislikes_ratio"] = round((fiveChannelsAnalysis["dislikes"] / fiveChannelsAnalysis["views"] ) * 100,2)
```

```python
fiveChannelsAnalysis.head()
```

```python
biini = np.linspace( fiveChannelsAnalysis["likes_ratio"].min() , fiveChannelsAnalysis["likes_ratio"].max(), 6)
```

```python
biini
```

```python
fiveChannelsAnalysis["Likeratio_range%"]  =  pd.cut(fiveChannelsAnalysis["likes_ratio"] , 
                                                bins= biini , 
                                                labels= ["0 to 2.9 % likes" , 
                                                         "3 to 5.9 % likes",
                                                         "6 to 9.9 % likes",
                                                         "10 to 12.9 % likes",
                                                         "13 to 18 % likes"])
```

```python
fiveChannelsAnalysis.head()
```

```python
px.pie(fiveChannelsAnalysis , 
       names="Likeratio_range%"  , 
       hole= 0.6 ,
       title= "figure 9.1: Likes Ratio of YouTube Videos" ,
       color_discrete_sequence= ["red"])
```

```python
bins = np.linspace( fiveChannelsAnalysis["dislikes_ratio"].min() , fiveChannelsAnalysis["dislikes_ratio"].max(), 6)
```

```python
bins
```

```python
fiveChannelsAnalysis["DisLikeRatio_range%"]  =  pd.cut(fiveChannelsAnalysis["dislikes_ratio"] , 
                                                bins= bins , 
                                                labels= ["0 to 0.8 % dislikes" , 
                                                         "0.9 to 1.8 % dislikes",
                                                         "1.9 to 2.8 % dislikes",
                                                         "2.9 to 3.8 % dislikes",
                                                         "3.9 to 5 % dislikes"])
```

```python
px.pie(fiveChannelsAnalysis , 
       names="DisLikeRatio_range%"  , 
       hole= 0.6 ,
       title= "figure 9.2: Dislikes Ratio of YouTube Videos" ,
       color_discrete_sequence= ["red"])
```
